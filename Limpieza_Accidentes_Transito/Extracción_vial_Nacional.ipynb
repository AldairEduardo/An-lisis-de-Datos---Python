```python
# Ruta del archivo PDF
pdf_file = "../Limpieza_Accidentes_Transito/pdf/RED VIAL NACIONAL.pdf"

# Lista para almacenar todas las tablas procesadas
tablas_procesadas = []
primer_encabezado = None  # Variable para almacenar los encabezados de la primera tabla

# Abrir el PDF con pdfplumber
with pdfplumber.open(pdf_file) as pdf:
    # Recorrer todas las páginas del PDF
    for i, page in enumerate(pdf.pages):
        print(f"Procesando página {i + 1}...")

        # Extraer todas las tablas de la página actual
        tablas = page.extract_tables()

        # Procesar cada tabla extraída
        for j, tabla in enumerate(tablas):
            print(f"Procesando tabla {j + 1} de la página {i + 1}...")

            # Convertir la tabla a un DataFrame
            df = pd.DataFrame(tabla)

            # Eliminar filas completamente vacías
            df.dropna(how="all", inplace=True)

            # Verificar si el DataFrame tiene al menos 3 columnas
            if len(df.columns) >= 3:
                # Seleccionar solo las columnas 0, 1 y 2
                df = df.iloc[:, [0, 1, 2]]

                # Tomar encabezados de la segunda fila (índice 1) de la primera tabla
                if primer_encabezado is None:
                    primer_encabezado = df.iloc[1].tolist()  # Guardar los encabezados
                    df.columns = primer_encabezado  # Usar los encabezados de la primera tabla
                    df = df[2:]  # Eliminar las filas 0 y 1 (títulos y encabezados originales)
                else:
                    # Asignar encabezados desde la primera tabla
                    df.columns = primer_encabezado

                    # Eliminar filas que coincidan exactamente con los encabezados
                    df = df[~df.eq(primer_encabezado).all(axis=1)]
                    
                # Eliminar filas completamente vacías después del procesamiento
                df.dropna(how="all", inplace=True)

                # Agregar la tabla procesada a la lista
                tablas_procesadas.append(df)
            else:
                print(f"Tabla {j + 1} de la página {i + 1} omitida (menos de 3 columnas).")

# Combinar todas las tablas en un único DataFrame
if len(tablas_procesadas) > 0:
    df_final = pd.concat(tablas_procesadas, ignore_index=True)

    # Exportar todas las tablas combinadas a un archivo Excel
    output_file = "../Limpieza_Accidentes_Transito/excel/RED_VIAL_NACIONAL_LIMPIO.xlsx"
    df_final.to_excel(output_file, index=False, engine='openpyxl')

    print(f"Archivo exportado exitosamente: {output_file}")
else:
    print("No se encontraron tablas válidas en el PDF.")

```
