```python
## Sumar los valores de las columnas 'FALLECIDOS' y 'HERIDOS'
total_fallecidos = df1.loc[df1['DEPARTAMENTO'] == 'N.I.', 'FALLECIDOS'].astype(int).sum()
total_heridos = df1.loc[df1['DEPARTAMENTO'] == 'N.I.', 'HERIDOS'].astype(int).sum()

print(f"Total de fallecidos: {total_fallecidos}")
print(f"Total de heridos: {total_heridos}")
```

```python
columnas = df1.columns[[2,3,4,5]]
fallecidos = []
heridos = []
for i in columnas:
    total_fallecidos = df1.loc[df1[i] == 'N.I.', 'FALLECIDOS'].astype(int).sum()    
    total_heridos = df1.loc[df1[i] == 'N.I.', 'HERIDOS'].astype(int).sum()
    fallecidos.append(total_fallecidos)
    heridos.append(total_heridos)

print(f'total de Fallecidos: {sum(fallecidos)}')
print(f'total de heridos: {sum(heridos)}')
```
